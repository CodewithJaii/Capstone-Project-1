<h3>Step 1: Load and Explore the Dataset</h3>


Load the dataset into a DataFrame and examine the structure of the data.


```python
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

#Create the dataframe
df = pd.read_csv(file_url)

#Display the top 10 records
df.head()
```

<h3>Step 2: Plot the Distribution of Industry</h3>


Explore how respondents are distributed across different industries.

- Plot a bar chart to visualize the distribution of respondents by industry.

- Highlight any notable trends.


```python
industry_counts = df['Industry'].value_counts()

# Display the value counts for reference
print("Distribution of Respondents Across Industries:")
print(industry_counts)

# Plot a bar chart to visualize the distribution
plt.figure(figsize=(12, 6))
industry_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Respondents Across Industries')
plt.xlabel('Industry')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Highlight notable trends
most_common_industry = industry_counts.idxmax()
print(f"\nThe most represented industry is: {most_common_industry}")
```

<h3>Step 3: Identify High Compensation Outliers</h3>


Identify respondents with extremely high yearly compensation.

- Calculate basic statistics (mean, median, and standard deviation) for `ConvertedCompYearly`.

- Identify compensation values exceeding a defined threshold (e.g., 3 standard deviations above the mean).


```python
mean_compensation = df['ConvertedCompYearly'].mean()
median_compensation = df['ConvertedCompYearly'].median()
std_dev_compensation = df['ConvertedCompYearly'].std()

# Define the threshold for high compensation (3 standard deviations above the mean)
threshold = mean_compensation + (3 * std_dev_compensation)

# Identify rows with extremely high compensation
high_compensation_outliers = df[df['ConvertedCompYearly'] > threshold]

# Print statistics and results
print(f"Mean Compensation: {mean_compensation}")
print(f"Median Compensation: {median_compensation}")
print(f"Standard Deviation: {std_dev_compensation}")
print(f"Compensation Threshold (3 SD above the mean): {threshold}")
print(f"Number of respondents with extremely high compensation: {len(high_compensation_outliers)}")
print("Examples of extremely high compensation outliers:")
print(high_compensation_outliers[['ConvertedCompYearly']].head())
```

<h3>Step 4: Detect Outliers in Compensation</h3>


Identify outliers in the `ConvertedCompYearly` column using the IQR method.

- Calculate the Interquartile Range (IQR).

- Determine the upper and lower bounds for outliers.

- Count and visualize outliers using a box plot.


```python
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
IQR = Q3 - Q1
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")

# Determine the upper and lower bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")

# Identify outliers
outliers = df[(df['ConvertedCompYearly'] < lower_bound) | (df['ConvertedCompYearly'] > upper_bound)]
print(f"Number of outliers: {len(outliers)}")

# Optionally, visualize the outliers using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, y='ConvertedCompYearly', color='skyblue')
plt.title('Box Plot of Yearly Compensation (Outliers Highlighted)')
plt.ylabel('Yearly Compensation')
plt.show()

# Count and display a few outliers
print("Outliers:")
print(outliers[['ConvertedCompYearly']].head())
```

<h3>Step 5: Remove Outliers and Create a New DataFrame</h3>


Remove outliers from the dataset.

- Create a new DataFrame excluding rows with outliers in `ConvertedCompYearly`.
- 
- Validate the size of the new DataFrame.


```python
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Create a new DataFrame excluding the outliers
df_no_outliers = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Validate the size of the new DataFrame
print("Original DataFrame size:", df.shape)
print("New DataFrame size (without outliers):", df_no_outliers.shape)
```

<h3>Step 6: Correlation Analysis</h3>


Analyze the correlation between `Age` (transformed) and other numerical columns.

- Map the `Age` column to approximate numeric values.

- Compute correlations between `Age` and other numeric variables.

- Visualize the correlation matrix.


```python
age_mapping = {
    "18-24 years": 21,
    "25-34 years": 29,
    "35-44 years": 39,
    "45-54 years": 49,
    "55-64 years": 59,
    "65 years or older": 70
}

df['Age_Numeric'] = df['Age'].map(age_mapping)

# Filter numeric columns only
numeric_cols = df.select_dtypes(include=['number'])

# Compute correlations between numeric columns
correlation_matrix = numeric_cols.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title("Correlation Matrix")
plt.show()
```
