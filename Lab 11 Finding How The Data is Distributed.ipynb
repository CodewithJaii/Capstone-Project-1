### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
# Step 3: Handle Missing Data

# Display missing values in the dataset
print("\nMissing Values Summary:")
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0]
print(missing_values)

# Handle missing values:
# For numeric columns - Replace missing values with the mean
for column in df.select_dtypes(include=['float64', 'int64']):
    if df[column].isnull().sum() > 0:
        df[column].fillna(df[column].mean(), inplace=True)

# For categorical columns - Replace missing values with the mode
for column in df.select_dtypes(include=['object']):
    if df[column].isnull().sum() > 0:
        df[column].fillna(df[column].mode()[0], inplace=True)

# Verify that all missing values are handled
print("\nMissing Values After Handling:")
print(df.isnull().sum().sum())  # Should print 0 if all missing values are handled

```
